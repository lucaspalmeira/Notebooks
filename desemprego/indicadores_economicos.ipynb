### Economia internacional - Indicadores de atividade econômica - Indicadores de oferta - Taxa de desemprego (ALE, CAN, EUA, FRA, ITA, JAP, UK).

Fonte: Banco central do Brasil - OCDE. 

Disponível em: https://www3.bcb.gov.br/sgspub/localizarseries/localizarSeries.do?method=prepararTelaLocalizarSeries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_al = pd.read_csv('desemprego_alemanha.csv')
df_ca = pd.read_csv('desemprego_canada.csv')
df_eu = pd.read_csv('desemprego_eua.csv')
df_fr = pd.read_csv('desemprego_franca.csv')
df_it = pd.read_csv('desemprego_italia.csv')
df_jp = pd.read_csv('desemprego_japao.csv')
df_uk = pd.read_csv('desemprego_uk.csv')
```

```python
for i in [df_al, df_ca, df_eu,
          df_fr, df_it, df_jp, df_uk]:
    i.drop(i.tail(1).index, inplace=True)
    i['data'] = [data[0:7] for data in list(i.index)]
    i['data'] = pd.to_datetime(i['data'], format='%m/%Y')
    x = list(i.index)
    x = [i[8:] for i in x]
    y = [j for j in i[list(i.columns)[0]]]
    y = [str(i) for i in y]
    percent = [valor_x + '.' + valor_y[:-2] for valor_x, valor_y in zip(x, y)]
    i['Taxa de desemprego %'] = percent
    i['Taxa de desemprego %'] = pd.to_numeric(i['Taxa de desemprego %'], errors='coerce')
```

```python
for i in [df_al, df_ca, df_eu,
          df_fr, df_it, df_jp, df_uk]:
    display(i)
```

## Visualização

```python
plt.style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=(30, 15))
ax.plot(df_al['data'], df_al['Taxa de desemprego %'], label='ALE')
ax.plot(df_ca['data'], df_ca['Taxa de desemprego %'], label='CAN')
ax.plot(df_eu['data'], df_eu['Taxa de desemprego %'], label='EUA')
ax.plot(df_fr['data'], df_fr['Taxa de desemprego %'], label='FRA')
ax.plot(df_it['data'], df_it['Taxa de desemprego %'], label='ITA')
ax.plot(df_jp['data'], df_jp['Taxa de desemprego %'], label='JAP')
ax.plot(df_uk['data'], df_uk['Taxa de desemprego %'], label='UK')
ax.set_xlabel('Ano', fontsize=20)
ax.set_ylabel('(%)', fontsize=20)
ax.set_title('Taxa de desemprego', fontsize=25)
ax.legend();
```
