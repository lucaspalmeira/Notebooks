```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('papel_moeda_emitido.csv')
df1 = pd.read_csv('base_monetaria_restrita.csv')
df2 = pd.read_csv('reservas_bancarias.csv')
```

## Separação da coluna em duas para cada data frame

```python
# BM - Papel moeda emitido
df[['data', 'Papel moeda emitido']] = df['Data;1780 - BM - Papel moeda emitido (saldo di�rio) - u.m.c. (mil)'].str.split(';', expand=True)
```

```python
# BM - Base monetária restrita 
df1[['data', 'Base monetaria restrita']] = df1['Data;1782 - BM - Base monet�ria restrita (saldo di�rio) - u.m.c. (mil)'].str.split(';', expand=True)
```

```python
# BM - Reservas bancárias
df2[['data', 'Reservas bancarias']] = df2['Data;1781 - BM - Reservas banc�rias (saldo di�rio) - u.m.c. (mil)'].str.split(';', expand=True)
```

Deleção da coluna em que os dados estão juntos

```python
# BM - Papel moeda emitido
# drop
df = df.drop(['Data;1780 - BM - Papel moeda emitido (saldo di�rio) - u.m.c. (mil)'], axis=1)
```

```python
# BM - Base monetária restrita 
# drop
df1 = df1.drop(['Data;1782 - BM - Base monet�ria restrita (saldo di�rio) - u.m.c. (mil)'], axis=1)
```

```python
# BM - Reservas bancárias
# drop
df2 = df2.drop(['Data;1781 - BM - Reservas banc�rias (saldo di�rio) - u.m.c. (mil)'], axis=1)
```

## Transformando os dados em inteiros ("int()")

```python
# BM - Papel moeda emitido

bm = []
for x in ['.']:
    for y in df['Papel moeda emitido']:
        y = y.replace(x, '')
        bm.append(y)

df['BM - Papel moeda emitido'] = bm
df = df.drop(6859)
df['BM - Papel moeda emitido'] = pd.to_numeric(df['BM - Papel moeda emitido'], errors='coerce')
df = df.drop('Papel moeda emitido', axis=1)
```

```python
# BM - Base monetária restrita 

bm = []
for x in ['.']:
    for y in df1['Base monetaria restrita']:
        y = y.replace(x, '')
        bm.append(y)
        
df1['BM - Base monetaria restrita'] = bm
df1 = df1.drop(6859)
df1['BM - Base monetaria restrita'] = pd.to_numeric(df1['BM - Base monetaria restrita'], errors='coerce')
df1 = df1.drop('Base monetaria restrita', axis=1)
```

```python
# BM - Reservas bancárias

bm = []
for x in ['.']:
    for y in df2['Reservas bancarias']:
        y = y.replace(x, '')
        bm.append(y)
        
df2['BM - Reservas bancarias'] = bm
df2 = df2.drop(6859)
df2['BM - Reservas bancarias'] = pd.to_numeric(df2['BM - Reservas bancarias'], errors='coerce')
df2 = df2.drop('Reservas bancarias', axis=1)
```

## Trasnformando a coluna data de formato object para datatime

```python
df['data'] = pd.to_datetime(df['data'], format='%d/%m/%Y')
```

```python
df1['data'] = pd.to_datetime(df1['data'], format='%d/%m/%Y')
```

```python
df2['data'] = pd.to_datetime(df2['data'], format='%d/%m/%Y')
```

## Visualização dos dados

```python
fig = plt.figure(figsize = (20, 10))
# figsize = (float, float)
plt.plot(df['data'], df['BM - Papel moeda emitido'])
plt.title('BM - Papel moeda emitido', fontsize=20)
plt.xlabel('Ano', fontsize=15)
plt.ylabel('Saldo diário - u.m.c. (mil)', fontsize=15)
plt.savefig('bm_papel_moeda_emitido.png', format='png')
plt.show()
```
